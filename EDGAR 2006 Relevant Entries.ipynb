```python
import numpy as np
import pandas as pd
```

```python
var_10K = "10-K"
var_10Q = "10-Q"
var_20F = "20-F"
var_40F = "40-F"
var_CF = "CF"
var_BS = "BS"
var_IS = "IS"

folder_pas = "common files"
folder_fel = "this folder has the quarterly and annual files"
folder_mai = "common files are here"
```

# Annual Files

```python
# Combine Annual Entries into a Single File

df_col_types = {"sic":str,"cik":str,"fy":str,"qtrs":str,"line":str}
df_all_10K = pd.DataFrame()
# Get all Entries
df_col_types = {"sic":str,"cik":str,"fy":str,"qtrs":str}
df_all_10K = pd.read_csv((folder_pas+"10-K_AllEntries.csv.zip"),dtype=df_col_types)
df_all_10K.drop(["Unnamed: 0"],axis=1,inplace=True)
df_all_10K["tag"] = df_all_10K["tag"].str.upper()
print(len(df_all_10K))

df_all_20F = pd.read_csv((folder_pas+"20-F_AllEntries.csv.zip"),dtype=df_col_types)
df_all_20F.drop(["Unnamed: 0"],axis=1,inplace=True)
df_all_20F["tag"] = df_all_20F["tag"].str.upper()
print(len(df_all_20F))

df_all_40F = pd.read_csv((folder_pas+"40-F_AllEntries.csv.zip"),dtype=df_col_types)
df_all_40F.drop(["Unnamed: 0"],axis=1,inplace=True)
df_all_40F["tag"] = df_all_40F["tag"].str.upper()
print(len(df_all_40F))

df_ANNUAL = df_all_10K.copy()
print(len(df_ANNUAL))
df_ANNUAL = df_ANNUAL.append(df_all_20F,ignore_index=True)
print(len(df_ANNUAL))
df_ANNUAL = df_ANNUAL.append(df_all_40F,ignore_index=True)
print(len(df_ANNUAL))

df_ANNUAL_temp = pd.DataFrame()
df_all_10K = pd.DataFrame()
df_all_20F = pd.DataFrame()
df_all_40F = pd.DataFrame()

df_ANNUAL["date_period"] = pd.to_datetime(df_ANNUAL["date_period"])
df_ANNUAL = df_ANNUAL[(~(df_ANNUAL["fy"].isnull()))]
df_ANNUAL["calyr"] = df_ANNUAL["date_period"].dt.year
# The field "fy" is incorrect for a few entries
df_ANNUAL["ffy"] = np.where(df_ANNUAL["fy"].astype(int) <= df_ANNUAL["calyr"],df_ANNUAL["calyr"],df_ANNUAL["fy"])
df_ANNUAL["cfy"] = df_ANNUAL["cik"] + "-" + df_ANNUAL["ffy"].astype(str)

df_ANNUAL.rename(columns={"date_period":"fdate"},inplace=True)
print(len(df_ANNUAL))
# Coreg shows the data of subsidiaries. Removing the entries.
# None that 10-K has entries from 10-Q as well.
df_ANNUALx = df_ANNUAL[["cik","name","fdate","qtrs","uom","value","ffy","cfy","tag","form","fp","line","stmt","negating","sic","countryba"]][df_ANNUAL["coreg"].isnull()]
df_ANNUALx.to_csv((folder_fel+"ANNUAL ALL.csv.zip"),compression="zip")
print(len(df_ANNUALx))
```

# Quarterly Files

```python
# Split the quarterly file into separate IS, BS and CF

# Get all Entries
df_col_types = {"sic":str,"cik":str,"fy":str,"qtrs":str}
df_all_10Qa = pd.read_csv((folder_pas+"10-Q_2009_2015_AllEntries.csv.zip"),dtype=df_col_types)
df_all_10Qa.drop(["Unnamed: 0"],axis=1,inplace=True)
df_all_10Qa["tag"] = df_all_10Qa["tag"].str.upper()
print(len(df_all_10Qa))

df_all_10Qx = df_all_10Qa[["cik","name","date_period","qtrs","uom","value","tag","form","fy","fp","line","stmt","negating","sic","countryba"]][df_all_10Qa["coreg"].isnull()]
print(len(df_all_10Qx))
df_all_10Qa = pd.DataFrame()

df_all_10Qb = pd.read_csv((folder_pas+"10-Q_2016_2020_AllEntries.csv.zip"),dtype=df_col_types)
df_all_10Qb.drop(["Unnamed: 0"],axis=1,inplace=True)
df_all_10Qb["tag"] = df_all_10Qb["tag"].str.upper()
print(len(df_all_10Qb))

df_all_10Qy = df_all_10Qb[["cik","name","date_period","qtrs","uom","value","tag","form","fy","fp","line","stmt","negating","sic","countryba"]][df_all_10Qb["coreg"].isnull()]
print(len(df_all_10Qy))
df_all_10Qb = pd.DataFrame()

# GET ALL TAGS
df_all_tags = pd.read_excel((folder_mai+"Tags and Stats.xlsx"),sheet_name="Tag List")
df_all_tags.sort_values(by=["stmt","metric","tag_order"],ascending=True,inplace=True)
print(len(df_all_tags))

df_bs_tags = df_all_tags[df_all_tags["stmt"] == var_BS]
df_is_tags = df_all_tags[df_all_tags["stmt"] == var_IS]
df_cf_tags = df_all_tags[df_all_tags["stmt"] == var_CF]
```

```python
df_QBSx = df_all_10Qx[(df_all_10Qx["stmt"] == var_BS)&(df_all_10Qx["qtrs"] == "0")&(df_all_10Qx["tag"].isin(df_bs_tags["tag"]))]
print(len(df_QBSx))

df_QBSy = df_all_10Qy[(df_all_10Qy["stmt"] == var_BS)&(df_all_10Qy["qtrs"] == "0")&(df_all_10Qy["tag"].isin(df_bs_tags["tag"]))]
print(len(df_QBSy))

df_QBS = df_QBSx.copy()
print(len(df_QBS))
df_QBSx = pd.DataFrame()
df_QBS = df_QBS.append(df_QBSy,ignore_index=True)
df_QBSy = pd.DataFrame()
print(len(df_QBS))

df_QISx = df_all_10Qx[(df_all_10Qx["stmt"] == var_IS)&(df_all_10Qx["qtrs"] == "1")&(df_all_10Qx["tag"].isin(df_is_tags["tag"]))]
print(len(df_QISx))

df_QISy = df_all_10Qy[(df_all_10Qy["stmt"] == var_IS)&(df_all_10Qy["qtrs"] == "1")&(df_all_10Qy["tag"].isin(df_is_tags["tag"]))]
print(len(df_QISy))

df_QIS = df_QISx.copy()
print(len(df_QIS))
df_QISx = pd.DataFrame()
df_QIS = df_QIS.append(df_QISy,ignore_index=True)
df_QISy = pd.DataFrame()
print(len(df_QIS))

df_QCFx = df_all_10Qx[(df_all_10Qx["stmt"] == var_CF)&(df_all_10Qx["tag"].isin(df_cf_tags["tag"]))]
print(len(df_QCFx))

df_QCFy = df_all_10Qy[(df_all_10Qy["stmt"] == var_CF)&(df_all_10Qy["tag"].isin(df_cf_tags["tag"]))]
print(len(df_QCFy))

df_QCF = df_QCFx.copy()
print(len(df_QCF))
df_QCFx = pd.DataFrame()
df_QCF = df_QCF.append(df_QCFy,ignore_index=True)
df_QCFy = pd.DataFrame()
print(len(df_QCF))

df_QCF["qtr_num"] = df_QCF["fp"].str[1:2]
df_QCF2 = df_QCF[df_QCF["qtr_num"] == df_QCF["qtrs"]]
print(len(df_QCF2))

df_QBS["date_period"] = pd.to_datetime(df_QBS["date_period"])
df_QBS["calyr"] = df_QBS["date_period"].dt.year
df_QBS["ffy"] = np.where(df_QBS["fy"].astype(int) <= df_QBS["calyr"],df_QBS["calyr"],df_QBS["fy"])
df_QBS["fyp"] = df_QBS["ffy"].astype(str) + "-" + df_QBS["fp"]
df_QBS["cfy"] = df_QBS["cik"] + "-" + df_QBS["ffy"].astype(str)
df_QBS["cfp"] = df_QBS["cik"] + "-" + df_QBS["ffy"].astype(str) + "-" + df_QBS["fp"]
df_QBS.rename(columns={"date_period":"fdate"},inplace=True)

df_QIS["date_period"] = pd.to_datetime(df_QIS["date_period"])
df_QIS["calyr"] = df_QIS["date_period"].dt.year
df_QIS["ffy"] = np.where(df_QIS["fy"].astype(int) <= df_QIS["calyr"],df_QIS["calyr"],df_QIS["fy"])
df_QIS["fyp"] = df_QIS["ffy"].astype(str) + "-" + df_QIS["fp"]
df_QIS["cfy"] = df_QIS["cik"] + "-" + df_QIS["ffy"].astype(str)
df_QIS["cfp"] = df_QIS["cik"] + "-" + df_QIS["ffy"].astype(str) + "-" + df_QIS["fp"]
df_QIS.rename(columns={"date_period":"fdate"},inplace=True)

df_QCF2["date_period"] = pd.to_datetime(df_QCF2["date_period"])
df_QCF2["calyr"] = df_QCF2["date_period"].dt.year
df_QCF2["ffy"] = np.where(df_QCF2["fy"].astype(int) <= df_QCF2["calyr"],df_QCF2["calyr"],df_QCF2["fy"])
df_QCF2["fyp"] = df_QCF2["ffy"].astype(str) + "-" + df_QCF2["fp"]
df_QCF2["cfy"] = df_QCF2["cik"] + "-" + df_QCF2["ffy"].astype(str)
df_QCF2["cfp"] = df_QCF2["cik"] + "-" + df_QCF2["ffy"].astype(str) + "-" + df_QCF2["fp"]
df_QCF2.rename(columns={"date_period":"fdate"},inplace=True)

df_QIS.to_csv((folder_fel+"10Q_INCOME.csv.zip"),compression="zip")
df_QBS.to_csv((folder_fel+"10Q_BALANC.csv.zip"),compression="zip")
df_QCF2.to_csv((folder_fel+"10Q_CASHFL.csv.zip"),compression="zip")
```

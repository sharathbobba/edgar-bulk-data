```python

df_col_types = {"sic":str,"cik":str,"cfy":str,"qtrs":str}
df_revenues_raw = pd.read_csv((folder_cos+"ANNUAL IS REVEN 1 FOUND.csv"),dtype=df_col_types)
df_revenues_raw.sort_values(by=["cik","cfy","stmt"],ascending=False,inplace=True)
df_revenues_raw.drop_duplicates(subset=["cik-cfy"],inplace=True)
print("Revenue Entries "+np.str(len(df_revenues_raw)))

df_col_types = {"sic":str,"cik":str,"cfy":str,"qtrs":str}
df_COS_raw = pd.read_csv((folder_cos+"ANNUAL IS COSTS 1 FOUND.csv"),dtype=df_col_types)
df_COS_raw.sort_values(by=["cik","cfy","stmt"],ascending=False,inplace=True)
df_COS_raw.drop_duplicates(subset=["cik-cfy"],inplace=True)
print("Cost of Sales Entries "+np.str(len(df_COS_raw)))

df_col_types = {"sic":str,"cik":str,"cfy":str,"qtrs":str}
df_GP_raw = pd.read_csv((folder_cos+"ANNUAL IS GROS MRGN 1 FOUND.csv"),dtype=df_col_types)
df_GP_raw.sort_values(by=["cik","cfy","stmt"],ascending=False,inplace=True)
df_GP_raw.drop_duplicates(subset=["cik-cfy"],inplace=True)
print("Gross Margin Entries "+np.str(len(df_GP_raw)))

df_col_types = {"sic":str,"cik":str,"cfy":str,"qtrs":str}
df_RD_raw = pd.read_csv((folder_cos+"ANNUAL IS RES DEV 1 FOUND.csv"),dtype=df_col_types)
df_RD_raw.sort_values(by=["cik","cfy","stmt"],ascending=False,inplace=True)
df_RD_raw.drop_duplicates(subset=["cik-cfy"],inplace=True)
print("Research/Development "+np.str(len(df_RD_raw)))

df_col_types = {"sic":str,"cik":str,"cfy":str,"qtrs":str}
df_SM_raw = pd.read_csv((folder_cos+"ANNUAL IS SEL MKT 1 FOUND.csv"),dtype=df_col_types)
df_SM_raw.sort_values(by=["cik","cfy","stmt"],ascending=False,inplace=True)
df_SM_raw.drop_duplicates(subset=["cik-cfy"],inplace=True)
print("Sales/Marketing Entries "+np.str(len(df_SM_raw)))

df_col_types = {"sic":str,"cik":str,"cfy":str,"qtrs":str}
df_OM_raw = pd.read_csv((folder_cos+"ANNUAL IS OPER MRGN 1 FOUND.csv"),dtype=df_col_types)
df_OM_raw.sort_values(by=["cik","cfy","stmt"],ascending=False,inplace=True)
df_OM_raw.drop_duplicates(subset=["cik-cfy"],inplace=True)
print("Operating Margin "+np.str(len(df_OM_raw)))

df_col_types = {"sic":str,"cik":str,"cfy":str,"qtrs":str}
df_NM_raw = pd.read_csv((folder_cos+"ANNUAL IS NET MRGN 1 FOUND.csv"),dtype=df_col_types)
df_NM_raw.sort_values(by=["cik","cfy","stmt"],ascending=False,inplace=True)
df_NM_raw.drop_duplicates(subset=["cik-cfy"],inplace=True)
print("Net Margin "+np.str(len(df_NM_raw)))

# Income Statement
df_REV = df_revenues_raw[["cik","name","sic","countryba","cfy","stmt",
                          "negating","date_period","uom","value Bn USD",
                          "FYQ","cik-cfy","Currency","Office","Industry Title"]]
df_REV.rename(inplace=True, columns={"value Bn USD":"Revenue"})

df_COS = df_COS_raw[["cik-cfy","value Bn USD"]]
df_COS.rename(inplace=True, columns={"value Bn USD":"Cost of Sales"})

df_GP = df_GP_raw[["cik-cfy","value Bn USD"]]
df_GP.rename(inplace=True, columns={"value Bn USD":"Gross Margin"})

df_RD = df_RD_raw[["cik-cfy","value Bn USD"]]
df_RD.rename(inplace=True, columns={"value Bn USD":"Research & Development"})

df_SM = df_SM_raw[["cik-cfy","value Bn USD"]]
df_SM.rename(inplace=True, columns={"value Bn USD":"Sales & Marketing"})

df_OM = df_OM_raw[["cik-cfy","value Bn USD"]]
df_OM.rename(inplace=True, columns={"value Bn USD":"Operating Margin"})

df_NM = df_NM_raw[["cik-cfy","value Bn USD"]]
df_NM.rename(inplace=True, columns={"value Bn USD":"Net Margin"})

print("Revenue Entries "+np.str(len(df_REV)))
print("Cost of Sales Entries "+np.str(len(df_COS)))
df_IS_COMB = pd.merge(df_REV,df_COS,on="cik-cfy",how="left")
print("Combined with COS "+np.str(len(df_IS_COMB)))

print("Gross Margin Entries "+np.str(len(df_GP)))
df_IS_COMB = pd.merge(df_IS_COMB,df_GP,on="cik-cfy",how="left")
print("Combined with GM "+np.str(len(df_IS_COMB)))

print("Sales/Marketing Entries "+np.str(len(df_SM)))
df_IS_COMB = pd.merge(df_IS_COMB,df_SM,on="cik-cfy",how="left")
print("Combined with SM "+np.str(len(df_IS_COMB)))

print("Research/Development "+np.str(len(df_RD)))
df_IS_COMB = pd.merge(df_IS_COMB,df_RD,on="cik-cfy",how="left")
print("Combined with RD "+np.str(len(df_IS_COMB)))

print("Operating Margin "+np.str(len(df_OM)))
df_IS_COMB = pd.merge(df_IS_COMB,df_OM,on="cik-cfy",how="left")
print("Combined with OM "+np.str(len(df_IS_COMB)))

print("Net Margin "+np.str(len(df_NM)))
df_IS_COMB = pd.merge(df_IS_COMB,df_NM,on="cik-cfy",how="left")
print("Combined with NM "+np.str(len(df_IS_COMB)))

df_IS_COMB["GM %"] = 100 * df_IS_COMB["Gross Margin"] / df_IS_COMB["Revenue"]
df_IS_COMB["OM %"] = 100 * df_IS_COMB["Operating Margin"] / df_IS_COMB["Revenue"]
df_IS_COMB["NM %"] = 100 * df_IS_COMB["Net Margin"] / df_IS_COMB["Revenue"]

df_IS_COMB["R&D %"] = 100 * df_IS_COMB["Research & Development"] / df_IS_COMB["Revenue"]
df_IS_COMB["S&M %"] =  100 * df_IS_COMB["Sales & Marketing"] / df_IS_COMB["Revenue"]
df_IS_COMB["R&D+S&M %"] = 100*(df_IS_COMB["R&D %"] + df_IS_COMB["S&M %"])
df_IS_COMB["R&D+S&M"] = (df_IS_COMB["Research & Development"] + df_IS_COMB["Sales & Marketing"])

df_IS_COMB["Cal GM"] = df_IS_COMB["Revenue"] - df_IS_COMB["Cost of Sales"]
df_IS_COMB["GM Diff"] = 100 * (df_IS_COMB["Cal GM"] - df_IS_COMB["Gross Margin"])/df_IS_COMB["Gross Margin"]

df_IS_COMB["Cal CoS"] = df_IS_COMB["Revenue"] - df_IS_COMB["Gross Margin"]
df_IS_COMB["CoS Diff"] = 100 * (df_IS_COMB["Cal CoS"] - df_IS_COMB["Cost of Sales"])/df_IS_COMB["Cost of Sales"]

df_IS_COMB["GM_"] = np.where(df_IS_COMB["Gross Margin"].isnull(),df_IS_COMB["Cal GM"],df_IS_COMB["Gross Margin"])
df_IS_COMB["GM_ %"] = 100 * (df_IS_COMB["GM_"] / df_IS_COMB["Revenue"])

df_IS_COMB["CoS_"] = np.where(df_IS_COMB["Cost of Sales"].isnull(),df_IS_COMB["Cal CoS"],df_IS_COMB["Cost of Sales"])
df_IS_COMB["CoS_ %"] = 100 * (df_IS_COMB["CoS_"] / df_IS_COMB["Revenue"])

df_IS_COMB.to_csv((folder_cos+"IS ANNUAL KPI.csv"))
print(len(df_IS_COMB))

df_IS_COMB_SAAS = pd.merge(df_IS_COMB,df_saas,on="cik",how="left")
df_IS_COMB_SAAS.to_csv((folder_cos+"IS SAAS ANNUAL KPI.csv"))
print(len(df_IS_COMB_SAAS))

df_FS_COMB = df_IS_COMB.copy()
print(len(df_IS_COMB_SAAS))
```
